**Loading the dependencies**

```python
import cv2
from ultralytics import YOLO
```

**Loading the pre-trained model**

```python
model = YOLO('yolov8n.pt')
```

**Loading the .yaml file**
**Epoch size=100**
**Image size=640**

```python
model.train(data='/kaggle/input/helmet-dataset/data.yaml', epochs=100, imgsz=640)
```

# **Testing of trained model on Web Cam**

**Loading the trained model**

```python
model = YOLO('/kaggle/working/runs/detect/train/weights/best.pt')
```

**Runtime video frames**

```python
cap = cv2.VideoCapture(0)
```

**Real time video camera loop for catching frames**

```python
while True:
    # Capture frame-by-frame
    ret, frame = cap.read()
    if not ret:
        print("Failed to grab frame")
        break

    # Perform inference on the frame
    results = model(frame)

    # Draw the results on the frame
    results.render()

    # Display the resulting frame
    cv2.imshow('Real-Time YOLOv8 Inference', results.ims[0])

    # Break the loop on 'q' key press
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
